# Question 1 &2 Write a python program which searches all the product under a particular product from www.amazon.in. The
product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for
guitars.

```python
driver=webdriver.Chrome()
```

```python
driver.get('http://www.amazon.in/')
```

```python
product=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
product.send_keys('laptop')
```

```python
search=driver.find_element(By.ID,"nav-search-submit-button")
search.click()
```

```python
availability=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div[2]/div/div[3]/span/div[1]/div/div/div[5]/ul[37]/span/span/li/span/a/span")
availability.click()
```

```python
url=[]
start=0
end=3
for page in range(start,end):
    url_path=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url_path:
        url.append(i.get_attribute('href'))
    next_button=driver.find_elements(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[23]/div/div/span/a[3]')
    
```

```python
 len(url)
```

```python
Names=[]
products=[]
prices=[]
returns_exchange=[]
Delivery=[]
Availability=[]
```

```python
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        brands=driver.find_element(By.XPATH,'//tr[@class="a-spacing-small po-brand"]/td[2]/span')
        Names.append(brands.text)
    except NoSuchElementException:
        Names.append('-')
        
    try:
        product_types=driver.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        products.append(product_types.text)
    except NoSuchElementException:
        products.append('-')   
        
    try:
        price_tags=driver.find_element(By.XPATH,'//span[@class="a-price a-text-price a-size-medium apexPriceToPay"]')
        prices.append(price_tags.text)
    except NoSuchElementException:
        prices.append('-')
        
    try:
        Returns=driver.find_element(By.XPATH,'//span[@class="a-declarative"]/div[2]')
        returns_exchange.append(Returns.text)
    except NoSuchElementException:
        returns_exchange.append('-')    
        
    try:
        Deliverys=driver.find_element(By.XPATH,'//div[@class="a-spacing-base"]/span/span')
        Delivery.append(Deliverys.text)
    except NoSuchElementException:
        Delivery.append('-')
        
    try:
        Availabilitys=driver.find_element(By.XPATH,'//div[@class="a-section a-spacing-none a-spacing-top-micro }"]/span')
        Availability.append(Availabilitys.text)
    except NoSuchElementException:
        Availability.append('-')     
```

```python
len(Names)
len(products)
len(prices)
len(returns_exchange)
len(Delivery)
len(Availability)
```

```python
df=pd.DataFrame({'name':Names,'Product':products,'Price':prices,'Return':returns_exchange,'Delivery':Delivery,'Availability':Availability,'Url':url})
df
```

```python
driver.close()
```

# 9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall
reviews, privates from price, dorms from price, facilities and property description.

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.hostelworld.com/')
```

```python
location=driver.find_element(By.CLASS_NAME,"native-input")
location.send_keys('London')
```

```python
search=driver.find_element(By.XPATH,'/html/body/div[3]/div/div[2]/main/header/div/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/ul/li[2]/button/div[2]/div[1]')
search.click()
```

```python
sear=driver.find_element(By.XPATH,'/html/body/div[3]/div/div[2]/main/header/div/div[2]/div[1]/div[1]/div/div[5]/button[2]/span')
sear.click()
```

```python
start=0
end=3
for page in range(start,end):
    name=[]
    for i in driver.find_elements(By.XPATH,'//div[@class="property-name"]/span'): 
        name.append(i.text)
        
    Distance=[]
    for i in driver.find_elements(By.XPATH,'//div[@class="property-distance"]/span[2]'): 
        Distance.append(i.text)
        
    review=[]
    for i in driver.find_elements(By.XPATH,'//div[@class="review"]/span'): 
        review.append(i.text)
        
    private_price=[]
    for i in driver.find_elements(By.XPATH,'//div[@class="property-accommodation-prices"]/div[1]'): 
        private_price.append(i.text.replace('\n',' '))
        
    dorms_price=[]
    for i in driver.find_elements(By.XPATH,'//div[@class="property-accommodation-price"][2]'): 
        dorms_price.append(i.text.replace('\n',' '))
        
    url=[]
    url_path=driver.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]')
    for i in url_path:
        url.append(i.get_attribute('href'))

next_button=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[1]/div/div/div[6]/div/button[2]')
next_button.click()        
```

```python
len(name),len(Distance),len(review),len(private_price),len(dorms_price),len(url)
```

```python
rating=[]
overall_review=[]
facilities=[]
property_description=[]
for i in url:
    driver.get(i)
    try:
        rate=driver.find_element(By.XPATH,'//div[@class="flex-20"]/div[1]/div[1]')
        rating.append(rate.text)
    except:
        rating.append('-')
        
    try:
        all_review=driver.find_element(By.XPATH,'//div[@class="flex-20"]/div[1]/div[2]/div[1]/span')
        overall_review.append(all_review.text)
    except:
        overall_review.append('-')
        
    try:
        facility=driver.find_element(By.XPATH,'//div[@class="facilities-container"]')
        facilities.append(facility.text)
    except:
        facilities.append('-')    
        
    try:
        description=driver.find_element(By.XPATH,'//div[@class="flex-80"]/div[6]/div/div[2]')
        property_description.append(description.text)
    except:
        property_description.append('-')    
```

```python
len(facilities),len(rating),len(overall_review),len(property_description)
```

```python
df=pd.DataFrame({'Name':name,'distance':Distance,'Review':review,'Rating':rating,'Overall_review':overall_review,'Private_price':private_price,'Dorms_price':dorms_price,'Facilities':facilities,'Description':property_description})
df
```

```python
driver.close()
```

# 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank
B) Name
C) Artist
D) Upload date
E) Views 

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
name=[]
artist=[]
upload_date=[]
views=[]

for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"][1]/tbody/tr/td[1]'):
    name.append(i.text)   
    
for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"][1]/tbody/tr/td[2]'):
    artist.append(i.text)   
    
for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"][1]/tbody/tr/td[4]'):
    upload_date.append(i.text)     
    
for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"][1]/tbody/tr/td[3]'):
    views.append(i.text)       
```

```python
len(name),len(artist),len(upload_date),len(views)
```

```python
df=pd.DataFrame({'Name':name,"Artist":artist,'Views':views,'Upload_date':upload_date})
df
```

```python
driver.close()
```

# 8. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/ You
have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year
 Note: - from the home page you have to go to the Show All Dataset page through code. 

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
datasets=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
datasets.click()
```

```python
url=[]
statr=0
end=65
for page in range(statr,end):
    url_path=driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/h2/a')
    for i in url_path:
        url.append(i.get_attribute('href'))
    next_buttons=driver.find_element(By.XPATH,'//div[@class="btn-group"]/button[2]')    
```

```python
len(url)
```

```python
name=[]
data_type=[]
task=[]
attribute_type=[]
instances=[]
no_of_attributes=[]
year=[]
for i in url:
    driver.get(i)
    try:
        names=driver.find_element(By.XPATH,'//div[@class="flex flex-col"][1]/div[1]/div[2]/div/h1')
        name.append(names.text)
    except NoSuchElementException:
        name.append('-')
        
    try:
        years=driver.find_element(By.XPATH,'//div[@class="flex flex-col"][1]/div[1]/div[2]/h2')
        year.append(years.text.split(' ')[2].split('/')[2])
    except NoSuchElementException:
        year.append('-')  
        
    try:
        dtype=driver.find_element(By.XPATH,'//div[@class="flex flex-col"][1]/div[2]//div[2]/div[1]/p')
        data_type.append(dtype.text)
    except NoSuchElementException:
        data_type.append('-')    
        
    try:
        tasks=driver.find_element(By.XPATH,'//div[@class="flex flex-col"][1]/div[2]//div[2]/div[3]/p')
        task.append(tasks.text)
    except NoSuchElementException:
        task.append('-')
    
    try:
        atype=driver.find_element(By.XPATH,'//div[@class="flex flex-col"][1]/div[2]//div[2]/div[4]/p')
        attribute_type.append(atype.text)
    except NoSuchElementException:
        attribute_type.append('-')
        
    try:
        instance=driver.find_element(By.XPATH,'//div[@class="flex flex-col"][1]/div[2]//div[2]/div[5]/p')
        instances.append(instance.text)
    except NoSuchElementException:
        instances.append('-')    
        
    try:
        n_attributes=driver.find_element(By.XPATH,'//div[@class="flex flex-col"][1]/div[2]//div[2]/div[6]/p')
        no_of_attributes.append(n_attributes.text)
    except NoSuchElementException:
        no_of_attributes.append('-') 
```

```python
len(name),len(data_type),len(task),len(attribute_type),len(instances),len(no_of_attributes),len(year)
```

```python
df=pd.DataFrame({'Name':name,'Data type':data_type,'Task':task,'Attribute type':attribute_type,'Instances':instances,'No of attributes':no_of_attributes,'Year':year})
df
```

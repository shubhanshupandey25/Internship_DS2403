# Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and
salary filter.web page https://www.naukri.com/

```python
location=driver.find_element(By.XPATH,'/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[4]/div[2]/div[2]/label/p/span[1]')
location.click()
```

# Q3: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link:
https://www.flipkart.com/apple-iphone-11-black-64-gb/productreviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=F
LIPKART

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART')
```

```python
Ratings=[]
Review_summarys=[]
Full_reviews=[]  
```

```python
start=0
end=10
for page in range(start,end):
    ratings=driver.find_elements(By.XPATH,'//div[@class="XQDdHH Ga3i8K"]')
    for i in ratings:
        Ratings.append(i.text)
        
    Review_sum=driver.find_elements(By.XPATH,'//p[@class="z9E0IG"]')
    for i in Review_sum:
        Review_summarys.append(i.text)
        
    Full_rev=driver.find_elements(By.XPATH,'//div[@class="ZmyHeo"]/div/div')
    for i in Full_rev:
        Full_reviews.append(i.text)
    next_button=driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[11]/span')
    next_button.click()
    time.sleep(2)
```

```python
len(Ratings),len(Review_summarys),len(Full_reviews)
```

```python
df=pd.DataFrame({'Rating':Ratings,'review_summary':Review_summarys,'Full_review':Full_reviews})
df
```

# Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps: webpagehttps://www.azquotes.com/

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.azquotes.com/')
```

```python
top_quotes=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/div/div[3]/ul/li[5]/a')
top_quotes.click()
```

```python
Quote=[]
Author=[]   
Quotes_type=[]  
```

```python
start=0
end=10
for page in range(start,end):
    quotes=driver.find_elements(By.XPATH,'//div[@class="wrap-block"]/p/a[2]')
    for i in quotes:
        Quote.append(i.text)
        
    authors=driver.find_elements(By.XPATH,'//div[@class="author"]/a')
    for i in authors:
        Author.append(i.text)
     
    types=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in types:
        Quotes_type.append(i.text)
    next_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a')
    next_button.click()
    time.sleep(3)
```

```python
len(Quote),len(Author),len(Quotes_type)
```

```python
df=pd.DataFrame({'quote':Quote,'author':Author,'quotes_type':Quotes_type})
df
```

# Q7: Write a python program to display list of respected former Prime Ministers of India (i.e. Name,
Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1')
```

```python
Names=[]
Born_dead=[]
Terms=[]
Remarks=[]
```

```python
names=driver.find_elements(By.XPATH,'//td[@valign="top"][2]/div')
for i in names:
    Names.append(i.text)
    
born=driver.find_elements(By.XPATH,'//td[@valign="top"][3]/div')
for i in born:
    Born_dead.append(i.text)
    
terms=driver.find_elements(By.XPATH,'//td[@valign="top"]/div[1]/span')
for i in terms:
    Terms.append(i.text)
    
remarks=driver.find_elements(By.XPATH,'//td[@valign="top"][5]/div')
for i in remarks:
    Remarks.append(i.text)
```

```python
len(Names),len(Born_dead),len(Terms),len(Remarks)
```

```python
df=pd.DataFrame({'name':Names,'Born_Dead':Born_dead,'Term':Terms,'Remarks':Remarks})
df
```
